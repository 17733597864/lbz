```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import make_pipeline

# 决策树不需要标准化，移除StandardScaler
tree = make_pipeline(DecisionTreeRegressor(random_state=42))

# 线性回归需要标准化，保留StandardScaler
line = make_pipeline(StandardScaler(), LinearRegression())

# 特征预处理管道
preprocessor = ColumnTransformer([
    # 数值特征：标准化
    ("nums", StandardScaler(),
     ['age', 'Medu', 'Fedu', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']),
    # 类别特征：独热编码（修正sparse参数为sparse_output）
    ('cats', OneHotEncoder(handle_unknown='ignore', sparse_output=False),
     ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'higher', 'internet', 'romantic'])
])



# 处理训练数据
X_train_processed = preprocessor.fit_transform(x_train)
X_test_processed = preprocessor.transform(x_test)


models = {
    "线性回归": LinearRegression(),
    "决策树": DecisionTreeRegressor(),

}

# 训练模型并计算各模型的MSE
for name, model in models.items():
    # 训练模型
    model.fit(X_train_processed, y_train)
    # 预测
    y_pred = model.predict(X_test_processed)
    # 计算MSE
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{name}的R²分数：{r2:.4f}")
    print(f"{name}的MSE：{mse:.4f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
a=x_train.loc[:,['age', 'Medu', 'Fedu', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']]
b=x_test.loc[:,['age', 'Medu', 'Fedu', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']]
a_ploy=poly_features.fit_transform(a)
b_ploy=poly_features.fit_transform(b)
lin_reg = LinearRegression()
lin_reg.fit(a_ploy, y_train)
y_pred = lin_reg.predict(b_ploy)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"线性回归的R²分数：{r2:.4f}")
print(f"线性回归的MSE：{mse:.4f}")
```

```python
y_train=data1.iloc[:,1:2]
x=data1.iloc[:,2:]
y_test=data3['Survived']
x,y_train,y_test
```

```python
from sklearn.impute import SimpleImputer

pip=make_pipeline(SimpleImputer(strategy='most_frequent'),StandardScaler())
pip2=make_pipeline(SimpleImputer(strategy='most_frequent'),OneHotEncoder(handle_unknown='ignore'))
preprocessor = ColumnTransformer([
    # 数值特征：标准化
    ("nums", pip,
     ['Pclass','Age','Fare']),
    # 类别特征：独热编码（修正sparse参数为sparse_output）
    ('cats', pip2,
     ['Sex','SibSp','Parch','Embarked']),
])
x_train_processed = preprocessor.fit_transform(x)
x_test_processed=preprocessor.transform(data2)
```

```python
models = {
    "线性回归": LinearRegression(),
    "决策树": DecisionTreeRegressor(),

}

# 训练模型并计算各模型的MSE
for name, model in models.items():
    # 训练模型
    model.fit(x_train_processed, y_train)
    # 预测
    y_pred = model.predict(x_test_processed)
    # 计算MSE
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{name}的R²分数：{r2:.4f}")
    print(f"{name}的MSE：{mse:.4f}")
```
